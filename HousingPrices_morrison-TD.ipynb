```python
# Load data 
train_data_csv = ('Resources/train.csv')
```

```python
train_data_df = pd.read_csv(train_data_csv)
train_data_df.tail(25)
```

```python
train_data_df.columns
```

```python
#Examine Missing Data

train_data_df.isnull()
```

```python
# Count NaN values in columns
train_data_df.isnull().sum()
```

```python
#Refer to Kaggle train data description for approriate values to
# to replace NaN values

train_data_df['LotFrontage'].fillna("0",inplace = True)
train_data_df['Alley'].fillna("None",inplace = True)
train_data_df['MSZoning'].fillna("None",inplace = True)
train_data_df['MasVnrType'].fillna("None",inplace = True)
train_data_df['MasVnrArea'].fillna("0",inplace = True)
train_data_df['BsmtQual'].fillna("NA",inplace = True)
train_data_df['BsmtCond'].fillna("NA",inplace = True)
train_data_df['BsmtExposure'].fillna("NA",inplace = True)
train_data_df['BsmtFinType1'].fillna("NA",inplace = True)
train_data_df['BsmtFinType2'].fillna("NA",inplace = True)
train_data_df['Electrical'].fillna("None",inplace = True)
train_data_df['FireplaceQu'].fillna("NO",inplace = True)
train_data_df['GarageType'].fillna("NA",inplace = True)
train_data_df['GarageYrBlt'].fillna("0",inplace = True)
train_data_df['GarageFinish'].fillna("NA",inplace = True)
train_data_df['GarageQual'].fillna("NA",inplace = True)
train_data_df['GarageCond'].fillna("NA",inplace = True)
train_data_df['PoolQC'].fillna("NA",inplace = True)
train_data_df['Fence'].fillna("NO",inplace = True)
train_data_df['MiscFeature'].fillna("NA",inplace = True)

```

```python
train_data_df.isnull().sum()
```

```python
#Create a total Sqft Column buy adding 1st,2nd, 
#basement if finished

#did not include lot Area, but can
train_data_df['TotalSQFT']= train_data_df['GrLivArea']+ train_data_df['BsmtFinSF2']+ train_data_df['BsmtFinSF1']
train_data_df['TotalSQFT']
```

```python
#Create a Price Per Square Foot Column
#Dividing the sales price by total square footage include living space size
train_data_df['PPsQFT']=train_data_df['SalePrice']/train_data_df['TotalSQFT']

train_data_df['PPsQFT']
```

```python
#Create a Price Per LotColumn
#Dividing the sales price by total square footage include lot size
train_data_df['PPLot']=train_data_df['SalePrice']/train_data_df['LotArea']

train_data_df['PPLot']
```
